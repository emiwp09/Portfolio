```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel
file_path = r"C:/Users/Emiliano/Desktop/SS python VSC/DatosFelkipeVadillo/Metadatos Eugenia V1.xlsx"
xls = pd.ExcelFile(file_path)
df = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_') #para que lea la hoa correcta 

# Contar los valores de desenlace del parto
desenlace_counts = df['desenlace_parto'].value_counts()

# Crear gráfico de barras
plt.figure(figsize=(8, 5))
desenlace_counts.plot(kind='bar', color=['blue', 'red'])

# Personalizar gráfico
plt.title('Distribución de Partos a Término y Pretérmino')
plt.xlabel('Tipo de Parto')
plt.ylabel('Cantidad de Casos')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Importar librerías necesarias
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo Excel nuevamente
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"
xls = pd.ExcelFile(file_path)

# Leer la hoja correcta
df1 = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Filtrar datos válidos para evitar valores nulos
df_filtered = df1[['peso_pregestacional_kg', 'desenlace_parto']].dropna()

# Contar la cantidad de partos por rango de peso y tipo de parto
peso_bins = pd.cut(df_filtered['peso_pregestacional_kg'], bins=10)  # Crear intervalos de peso
peso_parto_counts = df_filtered.groupby([peso_bins, 'desenlace_parto']).size().unstack()

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
peso_parto_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12,6))

# Personalización del gráfico
plt.title('Distribución del Tipo de Parto según el Peso Pregestacional', fontsize=14)
plt.xlabel('Peso Pregestacional (kg)', fontsize=12)
plt.ylabel('Cantidad de Partos', fontsize=12)
plt.xticks(rotation=45)  # Rotar etiquetas para mejor legibilidad
plt.legend(title="Tipo de Parto", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Mostrar gráfico
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el archivo Excel
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"
xls = pd.ExcelFile(file_path)

# Leer la hoja correcta
df1 = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Filtrar datos válidos para evitar valores nulos
df_filtered = df1[['edad_cronologicamujer', 'peso_pregestacional_kg', 'desenlace_parto']].dropna()

# Crear gráfico de dispersión con la edad en el eje X, el peso en el eje Y y el tipo de parto en colores
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_filtered, x='edad_cronologicamujer', y='peso_pregestacional_kg', hue='desenlace_parto', alpha=0.7, palette="rainbow")

# Personalización del gráfico
plt.title('Relación entre Edad, Peso Pregestacional y Tipo de Parto', fontsize=14)
plt.xlabel('Edad Cronológica de la Mujer (años)', fontsize=12)
plt.ylabel('Peso Pregestacional (kg)', fontsize=12)
plt.legend(title="Tipo de Parto")
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.show()
```
