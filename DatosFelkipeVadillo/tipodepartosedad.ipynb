```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo Excel
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"# Cambié "\" por "/" para evitar errores
xls = pd.ExcelFile(file_path)
df = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Contar los valores de la edad cronológica de la mujer y ordenarlos
edadcronologica_counts = df['edad_cronologicamujer'].value_counts().sort_index()


colores = plt.cm.viridis(np.linspace(0, 1, len(edadcronologica_counts)))# Crear colores diferentes para cada barra

# grafic
plt.figure(figsize=(10, 6))
edadcronologica_counts.plot(kind='bar', color=colores)
plt.title('Distribución de la Edad Cronológica de las Mujeres', fontsize=14)
plt.xlabel('Edad (años)', fontsize=12)
plt.ylabel('Cantidad de Casos', fontsize=12)
plt.xticks(rotation=45)  # Rotar etiquetas del eje X para mejor visibilidad
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo Excel
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"
  # Cambié "\" por "/" para evitar errores
xls = pd.ExcelFile(file_path)
df = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Seleccionar las columnas relacionadas con la edad
columnas_edad = ['sdg_visita', 'sdg_parto']
df_edades = df[columnas_edad].dropna()

# Crear histogramas para visualizar la distribución de edades
plt.figure(figsize=(10, 6))

# Histograma de semanas de gestación en la visita
plt.hist(df_edades['sdg_visita'], bins=20, alpha=0.7, label='Semanas Gestación (Visita)')

# Histograma de semanas de gestación al parto
plt.hist(df_edades['sdg_parto'], bins=20, alpha=0.7, label='Semanas Gestación (Parto)')


plt.title('Distribución de Edades en Diferentes Etapas')
plt.xlabel('Semanas')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
