```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo Excel
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"  # Cambié "\" por "/" para evitar errores
xls = pd.ExcelFile(file_path)
df = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Contar la cantidad de partos por edad y tipo de parto
edad_parto_counts = df.groupby(['edad_cronologicamujer', 'desenlace_parto']).size().unstack()

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
edad_parto_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12,6))

# Personalización del gráfico
plt.title('Distribución del Tipo de Parto según la Edad Cronológica', fontsize=14)
plt.xlabel('Edad de la Mujer (años)', fontsize=12)
plt.ylabel('Cantidad de Partos', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Tipo de Parto")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar gráfico
plt.show()
```

```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Cargar el archivo Excel
file_path = r"C:\Users\Emiliano\Desktop\SS python VSC\DatosFelkipeVadillo\Metadatos Eugenia V1.xlsx"  # Cambié "\" por "/" para evitar errores
xls = pd.ExcelFile(file_path)
df = xls.parse(sheet_name='cohorte_princesa_corregidaDESC_')

# Agrupar los datos por edad y peso pregestacional
edad_peso_counts = df.groupby(['edad_cronologicamujer', 'peso_pregestacional_kg']).size().unstack()

# Crear gráfico de barras
plt.figure(figsize=(12, 6))
ax = edad_peso_counts.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12,6))

# Personalización del gráfico
plt.title('Distribución del Peso Pregestacional según la Edad Cronológica', fontsize=14)
plt.xlabel('Edad de la Mujer (años)', fontsize=12)
plt.ylabel('Cantidad de Casos', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Peso Pregestacional (kg)", bbox_to_anchor=(1.05, 1), loc='upper left', ncol=2)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Ajustar la leyenda para que se vea mejor
plt.tight_layout()

# Mostrar gráfico
plt.show()


```
